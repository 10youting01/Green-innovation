```python
import os
import pandas as pd
import json
```

```python
import math
# get the number of words by log
def report_length(csr_text):
    length = len(csr_text)
    return length
```

```python
def process_all(base_paths):
    output_dir = "output_dataset"
    os.makedirs(output_dir, exist_ok=True)  # 確保輸出目錄存在
    output_path = os.path.join(output_dir, "report_length.csv")

    df = pd.DataFrame(columns=["folder_name", "market", "report_length"])  # 初始化 DataFrame

    for base_path in base_paths:  # 依序處理每個資料夾 (NASDAQ & NYSE)
        market_name = os.path.basename(base_path)  # 取得 NASDAQ / NYSE 作為標籤

        for folder_name in os.listdir(base_path):
            folder_path = os.path.join(base_path, folder_name)

            if not os.path.isdir(folder_path):  # 確保是資料夾
                continue

            json_path = os.path.join(folder_path, f"{folder_name}_v2_remove_punctuation.json")

            if os.path.exists(json_path):  # 檢查 JSON 檔案是否存在
                with open(json_path, "r", encoding="utf-8") as f:
                    csr_data = json.load(f)

                # 排除 key "1" 
                csr_text = "".join(value for key, value in csr_data.items() if key not in ("1"))

                # 計算長度
                length = report_length(csr_text)

                # 存入 DataFrame
                row_data = {"folder_name": folder_name, "market": market_name, "report_length": length}
                df = pd.concat([df, pd.DataFrame([row_data])], ignore_index=True)

                print(f"Processed: {folder_name} from {market_name}")
        # 儲存 CSV
    df.to_csv(output_path, index=False)
    print(f"Saved results to {output_path}")
```

```python
base_paths = [
    os.path.abspath("/home/francia/research_hub/csr_project/CSR_report_processed_v4/NASDAQ"),
    os.path.abspath("/home/francia/research_hub/csr_project/CSR_report_processed_v4/NYSE")
]
process_all(base_paths)
```
