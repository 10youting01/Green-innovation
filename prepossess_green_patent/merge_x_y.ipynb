```python
import pandas as pd
```

```python
csr = pd.read_csv("../output_dataset/csr_embeddings_leq2019_filled.csv")
gp = pd.read_csv("../Green_patent_dataset/merged_dataset/forward_count_value_ticker_all_y.csv")
```

```python
gp = gp[['year', 'ticker', 'exchange', 'total_5yr_forward_citations', 'patents_count', 'total_values_real']].copy()
gp['lagged'] = gp['year'] + 1
gp_ticker = gp[['ticker', 'exchange']].drop_duplicates()
gp
```

```python
csr[['exchange', 'ticker', 'year']] = csr['file_name'].str.extract(r'([^_]+)_([^_]+)_(\d+)$')
csr['year'] = csr['year'].astype(int)  # 將 year 轉為整數（可選）
csr
```

```python
# 複製 csr 的 year 當作合併用的 lagged 欄位
csr['lagged'] = csr['year']

# 進行合併：以 exchange, ticker, lagged 為鍵
csr_embedding_gp = csr.merge(gp, on=['exchange', 'ticker', 'lagged'], how='left')
csr_embedding_gp
```

```python
# 先取得 gp 中所有唯一的 exchange + ticker 組合
gp_ticker = gp[['exchange', 'ticker']].drop_duplicates()

# 合併篩選，只保留在 gp 中出現過的 exchange + ticker
filtered = csr_embedding_gp.merge(gp_ticker, on=['exchange', 'ticker'], how='inner')

# 檢查結果
print(filtered.shape)
print(filtered[['exchange', 'ticker']].drop_duplicates())
```
