```python
import pandas as pd
from typing import List

def filter_green_patents(files: List[str]) -> pd.DataFrame:
    """
    è®€å–å¤šä»½å¹´åº¦è³‡æ–™ï¼Œæ¸…ç†æ¬„ä½èˆ‡æ¢ä»¶éæ¿¾ï¼Œä¸¦é¡¯ç¤ºé€²åº¦èˆ‡æœ€çµ‚ä¿ç•™ç­†æ•¸ã€‚

    è™•ç†å…§å®¹ï¼š
    - åˆªé™¤æ¬„ä½åç¨±ä»¥ d_ é–‹é ­çš„æ¬„
    - ä¿ç•™å¾é–‹é ­åˆ° n_cpcï¼ˆå«ï¼‰ç‚ºæ­¢çš„æ¬„ä½
    - ç¯©é¸æ¢ä»¶ï¼š
        - cpc_sections åŒ…å« 'Y'
        - n_cpc ç‚º 2 æˆ– 4
        - assignee ä¸ç‚ºç©º
        - country ç‚º US
    """

    all_filtered = []
    total_records = 0

    for i, file in enumerate(files):
        try:
            print(f"ğŸ“„ ({i+1}/{len(files)}) æ­£åœ¨è™•ç†: {file}")

            df = pd.read_csv(file, dtype=str)

            # åˆªé™¤æ¬„ä½åç¨±ä»¥ d_ é–‹é ­çš„æ¬„
            df = df.loc[:, ~df.columns.str.startswith('d_')]
            df = df.drop(columns=['city', 'state', 'county'], errors='ignore')

            # ä¿ç•™å¾é–‹é ­åˆ° n_cpc æ¬„ä½ç‚ºæ­¢
            if 'n_cpc' in df.columns:
                n_cpc_index = df.columns.get_loc('n_cpc')
                df = df.iloc[:, :n_cpc_index + 1]  # ä¿ç•™åˆ° n_cpcï¼ˆå«ï¼‰
            else:
                print(f"âš ï¸ æ‰¾ä¸åˆ°æ¬„ä½ 'n_cpc'ï¼Œè·³éæ­¤æª”æ¡ˆ")
                continue

            # æ¢ä»¶éæ¿¾
            df_filtered = df[
                df['cpc_sections'].str.contains('Y', na=False) &
                df['n_cpc'].astype(str).isin(['2', '4']) &
                df['assignee'].notna() &
                (df['country'] == 'US')
            ]

            print(f"âœ… ç¯©é¸å¾Œä¿ç•™ {len(df_filtered)} ç­†")

            total_records += len(df_filtered)
            all_filtered.append(df_filtered)

        except Exception as e:
            print(f"âŒ ç„¡æ³•è™•ç† {file}: {e}")

    if all_filtered:
        result = pd.concat(all_filtered, ignore_index=True)
        print(f"\nğŸ‰ æ‰€æœ‰è³‡æ–™è™•ç†å®Œç•¢ï¼Œç¸½å…±ä¿ç•™ {total_records} ç­†ç¶ è‰²å°ˆåˆ©è³‡æ–™ã€‚")
        return result
    else:
        print("âš ï¸ æ²’æœ‰æˆåŠŸè™•ç†ä»»ä½•è³‡æ–™ã€‚")
        return pd.DataFrame()
```

```python
import glob

# å‡è¨­ä½ æ¯å¹´è³‡æ–™æ˜¯æ”¾åœ¨ä¸€å€‹è³‡æ–™å¤¾ä¸­ï¼Œä¾‹å¦‚ 'data/patents_*.csv'
files = sorted(glob.glob("/opt/hdd_1/research_hub/csr_project/Green_patent_dataset/patentView/*.csv"))  # è‡ªå‹•æŠ“å‡ºæ‰€æœ‰æª”æ¡ˆ

# åŸ·è¡Œåˆä½µèˆ‡éæ¿¾
green_patents_all = filter_green_patents(files)

# å„²å­˜çµæœï¼ˆå¯é¸ï¼‰
green_patents_all.to_csv("/opt/hdd_1/research_hub/csr_project/Green_patent_dataset/merged_dataset/filtered_green_patents_1997_2023.csv", index=False)
```
