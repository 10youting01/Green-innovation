## (2) 計算citation_patent_id的被引用次數: processed_patents_citations.csv

```python
import pandas as pd

file_path = 'g_us_patent_citation.tsv'
output_file = "processed_patents51to52.csv"  # 儲存處理後的檔案
chunk_size = 1000000  # 每批讀取 100 萬筆資料
start = 51
end = 52

application['patent_id'] = application['patent_id'].astype(str)
# 逐批讀取 TSV 檔案並處理
for i, chunk in enumerate(pd.read_csv(file_path, sep="\t", chunksize=chunk_size, low_memory=False, encoding="utf-8")):
    if i < start:
        print(i)
        continue
    elif i >= start and i <= end:
        print(i)
        chunk['patent_id'] = chunk['patent_id'].astype(str)
        # 有些日期年分異常
        chunk['citation_date'] = pd.to_datetime(chunk['citation_date'], errors='coerce')
        chunk = chunk[chunk['citation_date'].notna()]
        chunk['citation_patent_id'] = chunk['citation_patent_id'].astype(str)
        chunk['citation_year'] = chunk['citation_date'].dt.year
        # 計算每年被引用次數
        citation_counts = chunk.groupby(['citation_patent_id', 'citation_year']).size().reset_index(name='yearly_forward_citations')
        print(len(citation_counts))
        # 儲存處理後的資料，確保第一次儲存時包含表頭，後續的批次不包含表頭
        citation_counts.to_csv(output_file, mode='a', header=i == start, index=False)
    else:
        break
```

## version 1

```python
# 轉換日期欄位
patents_gvkey["filing_year"] = patents_gvkey["filing_year"].astype(int)
patents_gvkey["citation_year"] = patents_gvkey["citation_year"].astype(int)

# 確定 citation_year 是否在 filing_year 後五年內
patents_gvkey["within_5_years"] = (
    (patents_gvkey["citation_year"] >= patents_gvkey["filing_year"] + 1) & 
    (patents_gvkey["citation_year"] <= patents_gvkey["filing_year"] + 5)
)
# 計算 total_5yr_forward_citations
total_citations = (
    patents_gvkey[patents_gvkey["within_5_years"]] # 某個gvkey在某一filing_year中有幾個patent後五年有citations
    .groupby(["gvkey_numeric", "filing_year"])["yearly_forward_citations"]
    .sum()
    .reset_index()
)

# 找到 filing_year 後五年完全沒有 forward citations 的 gvkey_numeric
no_citations = (
    patents_gvkey.groupby(["gvkey_numeric", "filing_year"])["within_5_years"]
    .sum()
    .reset_index()
)

# 如果 filing_year 後 5 年內沒有 forward citations，就設為 0
final_result = pd.merge(total_citations, no_citations, on=["gvkey_numeric", "filing_year"], how="right")

# 如果 filing_year 後 5 年內沒有 forward citations，則 total_5yr_forward_citations 設為 0
final_result["yearly_forward_citations"].fillna(0, inplace=True)

# 保留需要的欄位
final_result.rename(columns={"yearly_forward_citations": "total_5yr_forward_citations"}, inplace=True)
final_result
```

## version 2

```python
# 可能要增加一個條件: 若 20 年間有至少發表過一個專利，他在其他沒有申請專利的年份就為 0

import pandas as pd
import numpy as np

# 假設 patents_gvkey 已經載入
patents_gvkey["filing_year"] = patents_gvkey["filing_year"].astype(int)
patents_gvkey["citation_year"] = patents_gvkey["citation_year"].astype(int)

# 確定 citation_year(被引用年分) 是否在 filing_year 後五年內 True/False
# 即該patent五年內有被引用過
patents_gvkey["within_5_years"] = (
    (patents_gvkey["citation_year"] >= patents_gvkey["filing_year"] + 1) & 
    (patents_gvkey["citation_year"] <= patents_gvkey["filing_year"] + 5)
)

# 計算 total_5yr_forward_citations (針對citation_year在filing_year後五年內的專利)
total_citations = (
    patents_gvkey[patents_gvkey["within_5_years"]] # 某個gvkey在某一filing_year中有幾個patent後五年有citations
    .groupby(["gvkey_numeric", "filing_year"])["yearly_forward_citations"]
    .sum()
    .reset_index()
)

# 找出每個gvkey在各個filing_year申請的所有專利中，有幾個patent在申請後五年有被其他專利引用
# within_5_years:某個gvkey在某一filing_year中有申請patent，其中有幾個patent在申請後五年有被其他專利引用
# True(該patent五年內有被引用過) 被視為 1，False 被視為 0
no_citations = (
    patents_gvkey.groupby(["gvkey_numeric", "filing_year"])["within_5_years"]
    .sum()
    .reset_index()
)

# 合併，確保 total_5yr_forward_citations 設為 0
final_result = pd.merge(total_citations, no_citations, on=["gvkey_numeric", "filing_year"], how="right")
# 合併後若yearly_forward_citations為空值，表示某公司該年所申請的所有專利在五年內沒有被引用過，所以設0
final_result["yearly_forward_citations"].fillna(0, inplace=True)

# 重新命名欄位
final_result.rename(columns={"yearly_forward_citations": "total_5yr_forward_citations"}, inplace=True)

# 確保 filing_year 2000-2019 全部存在
# 找出 2000-2019 至少有申請過專利的 gvkey
valid_gvkeys = patents_gvkey[(patents_gvkey["filing_year"] >= 2000) & (patents_gvkey["filing_year"] <= 2019)]["gvkey_numeric"].unique()

# 生成 2000-2019 的 filing_year
year_range = np.arange(2000, 2020)

# 建立所有 (gvkey_numeric, filing_year) 組合
all_combinations = pd.MultiIndex.from_product([valid_gvkeys, year_range], names=["gvkey_numeric", "filing_year"]).to_frame(index=False)

# 將原本的資料與補全的資料合併
final_result = pd.merge(all_combinations, final_result, on=["gvkey_numeric", "filing_year"], how="left")

# 缺失的 total_5yr_forward_citations 設為 0
final_result["total_5yr_forward_citations"].fillna(0, inplace=True)

# 輸出結果
final_result
# within_5_years
# 有值:某個gvkey在某一filing_year中有申請patent，其中有幾個patent在申請後五年有被其他專利引用
# 沒值:某個gvkey在某一filing_year中沒申請任何patent，forward_citations設0
```
