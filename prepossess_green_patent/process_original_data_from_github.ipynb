# Get Green patent from KPSS and cpc

```python
import pandas as pd
```

```python
KPSS = pd.read_csv("/opt/hdd_1/research_hub/csr_project/Green_patent_dataset/original_dataset/KPSS_2023.csv")
cpc = pd.read_csv("/opt/hdd_1/research_hub/csr_project/Green_patent_dataset/original_dataset/Match_patent_cpc_2023.csv")
```

```python
print(KPSS.columns)
print(cpc.columns)
```

```python
KPSS_cpc = KPSS.merge(cpc, how="left", left_on="patent_num", right_on="patent_num")
KPSS_cpc
```

```python
green_patent = KPSS_cpc[KPSS_cpc['cpc'].str.contains("Y02-|Y04S", na=False)]
green_patent
```

```python
green_patent['issue_date'] = green_patent['issue_date'].apply(
    lambda x: str(int(x))[:4] if pd.notnull(x) else None
)
green_patent['filing_date'] = green_patent['filing_date'].apply(
    lambda x: str(int(x))[:4] if pd.notnull(x) else None
)
green_patent.rename(columns={
    'issue_date': 'issue_year',
    'filing_date': 'filing_year'
}, inplace=True)
```

```python
green_patent.to_csv(
    "/opt/hdd_1/research_hub/csr_project/Green_patent_dataset/original_dataset/KPSS_2023_green_patent.csv",
    index=False
)
```
